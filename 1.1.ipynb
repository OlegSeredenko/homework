**4. –°–Ω–æ–≤–∞ –ø–æ–º–µ—Å—Ç–∏ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ `a`, `c` –∏ `f` —Å–æ–æ—Ç–≤–µ—Ç—Å–≤—É—é—â–∏–µ —Å–∏–º–≤–æ–ª—ã –∏–∑ —Å—Ç—Ä–æ–∫–∏ `s1`, –∞ –ø–æ—Ç–æ–º –≤—ã–≤–µ–¥–∏ –∑–Ω–∞—á–µ–Ω–∏—è —ç—Ç–∏—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö.**

```python
raw_text_1 = """
–ó–∞—è–≤–∫–∞ –Ω–∞ –æ–±—É—á–µ–Ω–∏–µ. –ö–æ–Ω—Ç–∞–∫—Ç–Ω–∞—è –ø–æ—á—Ç–∞ –∫–ª–∏–µ–Ω—Ç–∞: thebestcustomerever@skyeng.com –ò–º—è –∫–ª–∏–µ–Ω—Ç–∞: –ö–æ–Ω—Å—Ç–∞–Ω—Ç–∏–Ω –ö–æ–Ω—Å—Ç–∞–Ω—Ç–∏–Ω–æ–ø–æ–ª—å—Å–∫–∏–π. –û—á–µ–Ω—å –∂–¥–µ—Ç –æ—Ç–≤–µ—Ç–∞. 
"""

raw_text_2 = """
–ó–∞—è–≤–∫–∞ –Ω–∞ –æ–±—É—á–µ–Ω–∏–µ. –ê–ª–µ–∫—Å–µ–π –ê–ª–µ–∫—Å–µ–µ–≤–∏—á –ö–æ–Ω—Å—Ç–∞–Ω—Ç–∏–Ω–æ–ø–æ–ª—å—Å–∫–∏–π. –û—Å—Ç–∞–≤–∏–ª —Ä–æ–¥–∏—Ç–µ–ª—å, –æ–±—É—á–∞—Ç—å—Å—è, —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, –±—É–¥–µ—Ç —Ä–µ–±–µ–Ω–æ–∫.
–ö–æ–Ω—Ç–∞–∫—Ç–Ω–∞—è –ø–æ—á—Ç–∞ - thebestcustomer@skyeng.ru –ö–ª–∏–µ–Ω—Ç –ø–æ–∑–∏—Ç–∏–≤–µ–Ω, –∂–¥–µ—Ç —Å–≤—è–∑–∏ –≤ –±–ª–∏–∂–∞–π—à–µ–µ –≤—Ä–µ–º—è
"""


email = raw_text_2[raw_text_2.rfind(' ',0,raw_text_2.find('@'))+1:raw_text_2.find(' ',raw_text_2.find('@'),len(raw_text_2))]
print(email)
```

**11. –ü–æ—Å—á–∏—Ç–∞–π –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫ –≤ —Å–ø–∏—Å–∫–µ –Ω–∏–∂–µ –∏ –≤—ã–≤–µ–¥–∏ –æ—Ç–≤–µ—Ç –≤ –ø–æ–Ω—è—Ç–Ω–æ–º –≤–∏–¥–µ "–í –Ω–∞—à–µ–º —Å–ø–∏—Å–∫–µ __?__ –ø–æ–∫—É–ø–æ–∫".**

```python
s = '–ü—Ä–∏–≤–µ—Ç, –≤–æ—Ç —Ç–µ–±–µ —Å–ø–∏—Å–æ–∫ –ø–æ–∫—É–ø–æ–∫: –º–æ–ª–æ–∫–æ, —Ä–∏—Å, —Ö–ª–µ–±, –≥–æ—Ä–æ—Ö, –ø–µ—á–µ–Ω—å–µ, —Å—ã—Ä, —è–±–ª–æ–∫–∏, –∫–∞–ª—å–º–∞—Ä—ã.'
substr = s[s.find(':'):len(s)]
counter = substr.count(',')+1
print('–í –Ω–∞—à–µ–º —Å–ø–∏—Å–∫–µ '+str(counter)+' –ø–æ–∫—É–ø–æ–∫')
```
